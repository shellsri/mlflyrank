## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

**Lane:** Content Ranking

**ML Task Type:** Ranking

The objective is to rank content based on how relevant it is to a user's search query. Instead of simply deciding whether content is relevant or not, the model predicts a relevance score so that the most useful content appears at the top of search results.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

# Target or Proxy

Since true relevance is difficult to measure directly, I will use user engagement as a proxy.

Possible target variables include:
- Click (0 or 1)
- Click Through Rate (CTR)
- Time spent on page
- User rating

For this notebook, I will assume the target column is **clicked**, where:

- 1 = user clicked the content
- 0 = user did not click

## 3. Success metric

*One metric you can defend. What number means 'good'?*

# Success Metric

Since this is a ranking problem, appropriate evaluation metrics include:

- NDCG (Normalized Discounted Cumulative Gain)
- MAP (Mean Average Precision)

For simplicity, click prediction accuracy or AUC can also be monitored during experimentation.

The final goal is to maximize user engagement by showing more relevant content.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

# Unit of Analysis

One row represents one content item shown to one user for one search query.

Each row contains information such as:

- query
- content_id
- title
- category
- engagement features
- target (clicked)

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

# Why ML Beats a Fixed Rule

A fixed rule might always rank content using a simple measure like publication date or popularity.

However, users have different interests and search intents.

Machine learning can learn complex patterns from historical interactions, including:
- user preferences
- search intent
- content quality
- engagement history

This allows the ranking system to adapt over time and provide more personalized and relevant recommendations than static rules.
